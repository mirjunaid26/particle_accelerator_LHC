# Unsupervised ML Approach

```python
import torch
import numpy as np
import csv
import pandas as pd
import os 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
import itertools
```

```python
class SignalDetector():
    def __init__(self, src):
        """Loading Data"""
        print("Loading data", src)
        self.src = src
        self.items = os.listdir(self.src)
        self.dataset = pd.read_csv(os.path.join(self.src, self.items[0]))
        #print(self.dataset)
        self.normalize_analyze()

    """Plotting Confusion Matrix"""
    def plot_confusion_matrix(self, cm, classes,title='Confusion matrix', cmap=plt.cm.Blues):
        plt.imshow(cm, interpolation='nearest', cmap=cmap)
        plt.title(title)
        plt.colorbar()
        tick_marks = np.arange(len(classes))
        plt.xticks(tick_marks, classes, rotation=45)
        plt.yticks(tick_marks, classes)
        fmt = 'd'
        thresh = cm.max() / 2.
        for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
            plt.text(j, i, format(cm[i, j], fmt),
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")
        plt.tight_layout()
        plt.ylabel('True label')
        plt.xlabel('Predicted label')
        plt.show()

    """Signal Detection Function"""
    def normalize_analyze(self):
        #dataset = self.dataset.to_numpy()    
        self.dataset =self.dataset.fillna(0)  #Replace Nans by 0 
        X = self.dataset.iloc[:, :15]
        #X_std = (X - X.mean())/ (X.std() + 1e-8) #Standardization
        X_norm = (X - X.min())/(X.max() - X.min() + 1e-8)  #Min Max Normalization
        #print("Nan", np.where(np.isnan(X_norm)))
        
        #Visualyzing principal compoents
        from sklearn.decomposition import PCA
        pca = PCA(n_components = 14)
        pca.fit(X_norm)
        print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))

        ##Variance Proporation PC1 -> PC14
        print("pca_out.explained_variance_ratio_", pca.explained_variance_ratio_)
        loadings = pca.components_ 
        num_pc = pca.n_features_ 
        pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
        loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
        loadings_df['variable'] = X.columns.values
        loadings_df = loadings_df.set_index('variable')
        print("loadings")
        print(loadings_df.shape)
        #
        import seaborn as sns
        import matplotlib.pyplot as plt
        ax = sns.heatmap(loadings_df, annot=True, cmap='Spectral')
        plt.show()

        
        #Plotting CLass Distribution
        events = pd.value_counts(self.dataset['Class'], sort = True)
        events.plot(kind = 'bar')
        plt.title("Event Distribution")
        plt.xticks(range(2), ["Bg", "Signal"])
        plt.xlabel("Event Type")
        plt.ylabel("Freq")
        signal = self.dataset[self.dataset['Class'] == 1]
        bg = self.dataset[self.dataset['Class'] == 0]
        print(bg.shape, signal.shape)
        #plt.ylim((0, 15000))
        plt.show()

        #Analayzing Feature importance         
        fig, (axis1, axis2) = plt.subplots(2, 1, sharex=True)

        fig.suptitle('P1 Feature vs P14 Feature')
        axis1.scatter(signal.p3, signal.p14)
        axis1.set_title('Signal')
        axis2.scatter(bg.p3, bg.p14)
        axis2.set_title('Bg Event')
        plt.xlabel('P1 Feature')
        plt.ylabel('P14 Feature')
        plt.xlim((0, 23000))
        plt.show()

        f, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
        f.suptitle('Last Feature Value')
        bins = 15
        ax1.hist(signal.p14, bins = bins)
        ax1.set_title('Signal')
        ax2.hist(bg.p14, bins = bins)
        ax2.set_title('Bg')
        plt.xlabel('P 14')
        plt.ylabel('Value of P14 Feature')
        plt.xlim((0, 20000))
        plt.yscale('log')
        plt.show()

        self.dataset.hist(figsize=(15, 15), bins = 128)
        plt.show()


        #Preparaing Dataset
        self.dataset.sample(frac=1)  #Shuffle
        cols = self.dataset.columns.tolist()  #Extracting Labels
        trgt = cols[-1]
        cols = cols[:-1]

        Xtr = self.dataset.iloc[:280000, :-1]
        ytr = self.dataset.iloc[:280000, -1]
        Xt = self.dataset.iloc[280000:, :-1]
        yt = self.dataset.iloc[280000:, -1]

        print(Xtr.shape, ytr.shape, Xt.shape, yt.shape)
        from sklearn.neighbors import LocalOutlierFactor

        #Local Outlier for Signal from Large Background events 
        snb = 492/284315    #Signal to Bg Ratio
        model = LocalOutlierFactor(contamination=snb)
        """We are using Labels only for evaluation purposes here"""
        y_train_pred = model.fit_predict(Xtr)
        y_train_pred[y_train_pred == 1] = 0
        y_train_pred[y_train_pred == -1] = 1   #Replacing -1 by 1 and 1 by 0 to make it consistent with our class labels
        y_test_pred = model.fit_predict(Xt)
        y_test_pred[y_test_pred == 1] = 0
        y_test_pred[y_test_pred == -1] = 1


        #Evaluation Metrics
        from sklearn.metrics import confusion_matrix
        classes = np.array(['0','1'])
        cm_train = confusion_matrix(ytr, y_train_pred)
        self.plot_confusion_matrix(cm_train,["Bgnd", "Signal"])
        cm_test = confusion_matrix(yt, y_test_pred)
        self.plot_confusion_matrix(cm_test,["Bgnd", "Signal"])
```

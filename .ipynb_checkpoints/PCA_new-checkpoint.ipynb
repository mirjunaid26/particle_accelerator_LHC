#### **Data loading and pre-processing**

```python
data = pd.read_csv("new_physics_signal.csv")
df_columns = data.columns

data = data.dropna()
shuffled = data.sample(frac=1, random_state=4)
data_zero = shuffled.loc[shuffled['Class'] == 0].sample(n=800, random_state = 123)
data_one = shuffled.loc[shuffled['Class'] == 1]

data_sampled_concated = pd.concat([data_zero, data_one]).sample(frac=1, random_state=4)
data_sampled_concated = data_sampled_concated.reset_index(drop=True)
data = data_sampled_concated
```

```python
x_data = data.loc[:, data.columns[:-1]]
y_data = data.loc[:, data.columns[-1:]]
```

#### **Some utility functions**

```python
def scatter_plot_representations(x, colors):
    num_classes = len(np.unique(colors))
    palette = np.array(sns.color_palette("hls", num_classes))
    
    f = plt.figure(figsize=(8, 8))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40, c=palette[colors.astype(np.int)])
    
    plt.xlim(-25, 25)
    plt.ylim(-25, 25)
    ax.axis('tight')
    
    txts = []
    for i in range(num_classes):
        x_test, y_test = np.median(x[colors == i, :], axis=0)
        txt = ax.text(x_test, y_test, str(i), fontsize = 24)
        txt.set_path_effects([PathEffects.Stroke(linewidth=5, foreground="w"), PathEffects.Normal()])
        txts.append(txt)
    return f, ax, sc, txts
```

```python
def scatter_3d_plot(x, y):
    """
    x : pd.DataFrame
    y : list
    """
    ax = plt.figure(figsize=(16,10)).gca(projection='3d')
    cols = x.columns
    ax.scatter(
        xs = x[cols[0]], 
        ys = x[cols[1]], 
        zs = x[cols[2]], 
        c  = y, 
        cmap='tab10')
    ax.set_xlabel('pca component one')
    ax.set_ylabel('pca component two')
    ax.set_zlabel('pca component three')
    
    plt.legend([f'class {i}' for i in np.unique(y)], loc="lower right")
    plt.show()
    
    
```

### **PCA**

**PCA** is a linear feature extraction technique. It does a linear mapping of a high dimensional spatial data to a low dimensional space, such that the variance of the mapped low 
dimensional representation is maximized. So PCA drops the least important features and try to map those into the required components, which are obtained by calculating the maximum 
eigen vectors (got from the corresponding max eigen values) or the principle components tries to reconstruct the data into those pricniple component axes.

```python
pca = PCA(n_components=5)
results_from_pca = pca.fit_transform(x_data)
```

```python
print(f"Before transformation the shape of the original data: {x_data.shape} and after transformation from PCA the shape {results_from_pca.shape}")
```

```python
def get_pca_for_scaled_data(data, y , scale_type, df_columns = df_columns, n_components = 5, components_to_viz = 2):
    """
    data : pd.DataFrame
    y : list
    components : 2 or 3
    
    available options:
    1. None : This will not scale the data 
    2. minmax : This will do the MinMax scaler
    3. standard : This will perform standard scaling
    4. robust : This will do robust scaling 
    """
    
    if scale_type is None:
        data = data
    elif scale_type == "minmax":
        scaler = MinMaxScaler()
        data = scaler.fit_transform(data)
    elif scale_type == "standard":
        scaler = StandardScaler()
        data = scaler.fit_transform(data)
    else:
        scaler = RobustScaler()
        data = scaler.fit_transform(data)
    
    data = pd.DataFrame(data, columns = df_columns)
    
    pca = PCA(n_components=n_components)
    pca_results = pca.fit_transform(data)
    
    pca_columns = [f'pca_comp{i+1}' for i in range(n_components)]
    pca_df = pd.DataFrame(columns = pca_columns)
    
    for i in range(5):
        pca_df[f"pca_comp{i+1}"] = results_from_pca[:, i]
        
    top_k = sorted(zip(pca_columns, pca.explained_variance_ratio_), reverse=True)[:components_to_viz]
    top_k_cols = [comp[0] for comp in top_k]
    print(f"The top {components_to_viz} components : {top_k_cols} with max variance of {sorted(pca.explained_variance_ratio_, reverse=True)[:components_to_viz]}")
          
    top_k_components = pca_df.loc[:, top_k_cols]
    
    if components_to_viz == 2:
        scatter_plot_representations(top_k_components.values, y)
    else:
        scatter_3d_plot(top_k_components, y)
```

```python
data_for_pca = x_data
y_for_pca = y_data.values.reshape(len(x_data))

get_pca_for_scaled_data(data_for_pca, y_for_pca, scale_type=None, n_components=2, df_columns=df_columns[:-1], components_to_viz=2)
get_pca_for_scaled_data(data_for_pca, y_for_pca, scale_type=None, n_components=3, df_columns=df_columns[:-1], components_to_viz=3)
```
